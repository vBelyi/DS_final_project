```python
# Ініціалізація початкових рейтингів для кожної команди (може бути будь-якими початковими значеннями)
initial_rating = 1500  # Початковий рейтинг Ело

# Створення словника для збереження рейтингів кожної команди
team_ratings = {team: initial_rating for team in team_data['Team']}

# Функція для розрахунку очікуваного результату матчу між командами A і B
def expected_result(rating_A, rating_B):
    return 1 / (1 + 10 ** ((rating_B - rating_A) / 400))

# Проходження через кожен матч і оновлення рейтингів
for index, row in final_df_2019_2023.iterrows():
    home_team = row['HomeTeam']
    away_team = row['AwayTeam']
    result = row['FTR']  # Результат матчу: 'H' - перемога вдома, 'A' - перемога на виїзді, 'D' - нічия
    
    # Отримання поточних рейтингів команд
    rating_home = team_ratings[home_team]
    rating_away = team_ratings[away_team]
    
    # Розрахунок очікуваного результату для кожної команди
    expected_home = expected_result(rating_home, rating_away)
    expected_away = expected_result(rating_away, rating_home)
    
    # Оновлення рейтингів команд на основі фактичного результату
    if result == 'H':
        team_ratings[home_team] = rating_home + 32 * (1 - expected_home)
        team_ratings[away_team] = rating_away + 32 * (0 - expected_away)
    elif result == 'A':
        team_ratings[home_team] = rating_home + 32 * (0 - expected_home)
        team_ratings[away_team] = rating_away + 32 * (1 - expected_away)
    else:  # 'D' - нічия
        team_ratings[home_team] = rating_home + 32 * (0.5 - expected_home)
        team_ratings[away_team] = rating_away + 32 * (0.5 - expected_away)

# Створення нового датасету з рейтингами команд
ratings_ELO_df = pd.DataFrame({'Team': list(team_ratings.keys()), 'Elo_Rating': list(team_ratings.values())})
```

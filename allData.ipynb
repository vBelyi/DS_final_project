```python
import pandas as pd
```

```python
#Завантажимо всі файли
file2022_2023 = pd.read_csv('SeriaAs2022_2023.csv')
file2021_2022 = pd.read_csv('SeriaAs2021_2022.csv')
file2020_2021 = pd.read_csv('SeriaAs2020_2021.csv')
file2019_2020 = pd.read_csv('SeriaAs2019_2020.csv')
```

```python
#Робимо один датасет з усіх
df_2019_2023 = pd.concat([file2022_2023, file2021_2022, file2020_2021, file2019_2020], ignore_index = True)
```

```python
df_2019_2023
```

```python
#Тепер почистимо датасет від тих колонок, які нам не потрібні.
final_df_2019_2023 = df_2019_2023[['HomeTeam', 'AwayTeam', 'FTHG', 'FTAG', 'FTR', 'HS', 'AS', 'HST', 'AST', 'HC', 'AC', 'HF', 'AF', 'HY', 'AY', 'HR', 'AR']]
```

```python
final_df_2019_2023
```

```python
#Додамо ще 2 колонки, по набраним очкам. Де HTP - home team points, а ATP - away team points.
#Класично в футболі, за перемогу - 3 бали, нічию - по 1 балу, поразка - 0 балів.
final_df_2019_2023['HTP'] = final_df_2019_2023.apply(lambda row: 1 if row['FTR'] == 'D' else 3 if row['FTR'] == 'H' else 0, axis=1)
final_df_2019_2023['ATP'] = final_df_2019_2023.apply(lambda row: 1 if row['FTR'] == 'D' else 3 if row['FTR'] == 'A' else 0, axis=1)
```
